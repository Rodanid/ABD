```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python

df_main = pd.read_csv('E:/Downloads/stores-dist.csv')
df_main.head()
```

```python
df_main = df_main.rename(columns={'annual net sales':'sales',
                                  'number of stores in district':'stores'})
df_main.head()
```

```python
df_corr = df_main.corr()
df_corr
```

```python
df = df_main.drop('district', axis=1)
df.head()
```

```python
y = df['sales']
x = df.stores
```

```python
%matplotlib inline

plt.figure(figsize = (20,10))
plt.plot(x,y,'o',markersize=30)
plt.ylabel('Annual Net Sales', fontsize = 30)
plt.xlabel('Number of Stores in the District', fontsize = 30)
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)

plt.show()
```

```python
m,b = np.polyfit(x,y,1)
print('The slope of line is {:.2f}.'.format(m))
print('The y-inetcept is {:.2f}.'.format(b))
print('The besst fit =simple linear regression line is {:.2}x + {:.2f}'.format(m,b))
```

```python
y_mean = y.mean()
x_mean = x.mean()
print('The centroid for this dataset is x = {:.2f} and y = {:.2f}'.format(x_mean,y_mean))
```

```python
%matplotlib inline
plt.figure(figsize=(20,10))
plt.plot(x,y,'o',markersize = 14, label = 'Annual Net Sales')
plt.plot(x_mean,y_mean, '*', markersize = 30, color = 'r')
plt.plot(x, m*x+b, '-', label = 'Simple Linear Regression Line', linewidth = 4)
plt.ylabel('Annual Net Sales', fontsize = 30)
plt.xlabel('Number of Stores ifn District', fontsize = 30)
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
plt.annotate('Centroid', xy=(x_mean-0.1,y_mean-5),xytext=(x_mean-3,y_mean-20), arrowprops=dict(facecolor='black',shrink=0.05),fontsize=30)
plt.legend(loc = 'upper right', fontsize = 20)
```

```python
def predict(query):
    if query >=1:
        predict = m*query+b
        return predict
    else:
        print('You must have at least 1 store in the district to pedict the annual net sales.')
```
